```python
# Ans 7

import csv
from collections import Counter

# Load your dataset
data = []
with open(r"C:\Users\Anshu Kumari\Desktop\India Agriculture Crop Production.csv", 'r') as file:
    reader = csv.reader(file)
    next(reader) 
    for row in reader:
        try:
            value = float(row[5])  # Convert to float instead of int
            data.append(value)
        except (ValueError, IndexError):
            
            pass

# Calculate Mean
def cal_mean(data):
    total = sum(data)
    mean = total / len(data)
    return mean
mean = cal_mean(data)

# Calculate Median
def cal_median(data):
    sorted_data = sorted(data)
    n = len(data)
    
    if n % 2 == 0:
        median = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
    else:
        median = sorted_data[n // 2]
    return median
median = cal_median(data)

# Calculate Mode
def cal_mode(data):
     counts = Counter(data)
     mode = [k for k, v in counts.items() if v == max(counts.values())]
     return mode
mode = cal_mode(data)

# Print the mean, median, and mode values
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
```
